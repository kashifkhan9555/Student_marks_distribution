```python
# prompt: import all the library for data analysis

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statistics
```

### Drop unnamed Column

```python
# prompt: drop "unnamed colum"

df.drop(columns=['Unnamed: 0'], axis=1, inplace=True)
```

```python
df.head()
```

```python
# prompt: drop null values

df.dropna(axis=0, inplace=True)
df.isnull().sum()
```

```python
# prompt: use seaborn count function

ax = sns.countplot(df['Gender'])
ax.bar_label(ax.containers[0])
plt.figure(figsize=(8,8))
```

### from the above chart we observe that the females are more as compare to males.

```python
gb =df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb)
```

```python
# prompt: plot this on chart and also use heat map

df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'}).plot(kind="bar")
plt.figure(figsize=(8,8))
sns.heatmap(df.corr(), annot=True)
```

from the above chart we observe that the parents edu. has impact on the students marks

```python
gb1 =df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb1)
```

```python
# prompt: plot this on bar chart and also on heatmap

df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'}).plot(kind="bar")
plt.figure(figsize=(8,8))
sns.heatmap(df.corr(), annot=True)
```

```python
# prompt: plot box plot

df.boxplot(column=['MathScore', 'ReadingScore', 'WritingScore'], by='Gender')
plt.show()
```

```python
# prompt: fix outliers and show

df = df[df['MathScore'] < 200]
df = df[df['ReadingScore'] < 200]
df = df[df['WritingScore'] < 200]

df.boxplot(column=['MathScore', 'ReadingScore', 'WritingScore'], by='Gender')
plt.show()
```
